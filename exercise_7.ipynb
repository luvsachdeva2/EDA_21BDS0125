```python
#21BDS0125
#Luv Sachdeva
import pandas as pd

url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/horse.csv'
horse_df = pd.read_csv(url)
```

```python
dimensions = horse_df.shape

dimensions

```

```python
summary = horse_df.describe(include='all')
summary
```

```python
missing_data = horse_df.isnull().sum()
missing_data




```

```python
data_types = horse_df.dtypes
```

```python
head = horse_df.head()
head
```

```python
numerical_cols = horse_df.select_dtypes(include=['float64', 'int64']).columns
horse_df[numerical_cols] = horse_df[numerical_cols].fillna(horse_df[numerical_cols].median())

```

```python
categorical_cols = horse_df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    horse_df[col] = horse_df[col].fillna(horse_df[col].mode()[0])

```

```python
missing_data_after_cleaning = horse_df.isnull().sum().sum()

missing_data_after_cleaning
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='surgery', data=horse_df)
plt.title('Distribution of Surgery Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='age', data=horse_df)
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(horse_df['rectal_temp'], kde=True, bins=10)
plt.title('Distribution of Rectal Temperature')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(horse_df['pulse'], kde=True, bins=10)
plt.title('Distribution of Pulse')
plt.show()

```

```python
numerical_cols = horse_df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = horse_df.select_dtypes(include=['object']).columns


for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(horse_df[col], kde=True, bins=10)
    plt.title(f'Distribution of {col}')
    plt.show()


for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=horse_df)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```
