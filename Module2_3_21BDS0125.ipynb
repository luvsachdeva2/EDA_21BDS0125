```python
#21BDS0125
#Luv Sachdeva
import pandas as pd

url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/horse.csv'
horse_df = pd.read_csv(url)
```

```python
dimensions = horse_df.shape

dimensions

```

```python
summary = horse_df.describe(include='all')
summary
```

```python
missing_data = horse_df.isnull().sum()
missing_data




```

```python
data_types = horse_df.dtypes
```

```python
head = horse_df.head()
head
```

```python
numerical_cols = horse_df.select_dtypes(include=['float64', 'int64']).columns
horse_df[numerical_cols] = horse_df[numerical_cols].fillna(horse_df[numerical_cols].median())

```

```python
categorical_cols = horse_df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    horse_df[col] = horse_df[col].fillna(horse_df[col].mode()[0])

```

```python
missing_data_after_cleaning = horse_df.isnull().sum().sum()

missing_data_after_cleaning
```

```python
horse_df_dedup = horse_df.drop_duplicates()
print("Dataset Dimensions after Deduplication:", horse_df_dedup.shape)
```

```python
horse_df_dedup['temp_category'] = pd.cut(horse_df_dedup['rectal_temp'],
                                         bins=[35, 37, 39, 41],
                                         labels=['low', 'normal', 'high'])
horse_df_dedup['temp_category']
```

```python
horse_df_dedup['pulse_category'] = pd.cut(horse_df_dedup['pulse'],
                                          bins=[0, 60, 100, 200],
                                          labels=['low', 'normal', 'high'])
horse_df_dedup['pulse_category']
```

```python
horse_df_dedup['temp_quantile_bin'] = pd.qcut(horse_df_dedup['rectal_temp'],
                                              q=4,  # Create 4 quantile-based bins (quartiles)
                                              labels=['Q1', 'Q2', 'Q3', 'Q4'])
horse_df_dedup['temp_quantile_bin']
```

```python
horse_df_dedup['pulse_quantile_bin'] = pd.qcut(horse_df_dedup['pulse'],
                                               q=4,  # Create 4 quantile-based bins (quartiles)
                                               labels=['Q1', 'Q2', 'Q3', 'Q4'])
horse_df_dedup['pulse_quantile_bin']
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/horse.csv'
horse_df = pd.read_csv(url)
```

```python
date_rng = pd.date_range(start='2022-01-01', periods=len(horse_df), freq='D')
horse_df['date'] = date_rng

# Set the 'date' column as the index for Time Series Analysis
horse_df.set_index('date', inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(horse_df.index, horse_df['rectal_temp'], label='Rectal Temp', color='blue')
plt.plot(horse_df.index, horse_df['pulse'], label='Pulse', color='green')
plt.plot(horse_df.index, horse_df['respiratory_rate'], label='Respiratory Rate', color='red')
plt.title('Line Plot: Time Series of Rectal Temp, Pulse, and Respiratory Rate')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(horse_df.index, horse_df['rectal_temp'], label='Rectal Temp', color='blue', alpha=0.5)
plt.scatter(horse_df.index, horse_df['pulse'], label='Pulse', color='green', alpha=0.5)
plt.title('Scatter Plot: Rectal Temp and Pulse Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(horse_df.index, horse_df['rectal_temp'], label='Rectal Temp', color='blue', alpha=0.5)
plt.scatter(horse_df.index, horse_df['respiratory_rate'], label='Respiratory Rate', color='red', alpha=0.5)
plt.title('Scatter Plot: Rectal Temp and Respiratory Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
horse_df['week'] = horse_df.index.isocalendar().week
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='week', y='rectal_temp', data=horse_df)
plt.title('Box Plot: Weekly Rectal Temp Distribution')
plt.xlabel('Week Number')
plt.ylabel('Rectal Temp')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='week', y='pulse', data=horse_df)
plt.title('Box Plot: Weekly Pulse Distribution')
plt.xlabel('Week Number')
plt.ylabel('Pulse')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='week', y='respiratory_rate', data=horse_df)
plt.title('Box Plot: Weekly Respiratory Rate Distribution')
plt.xlabel('Week Number')
plt.ylabel('Respiratory Rate')
plt.xticks(rotation=45)
plt.show()
```
